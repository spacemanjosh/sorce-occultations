```python
import matplotlib.pyplot as plt
import scipy
import numpy as np
import math
import occultation_package as occ
import astropy
```

```python
def Average(lst):
    return sum(lst) / len(lst)
```

```python
def CD_Temp(Temp):
    z = 200000
    z_0 = 95000
    m = 0.02896
    g = 9.81
    k = 8.314462618
    N_0 = 1e18
    N = (N_0)*(np.power(math.e, ((-(z-(z_0))*(m*g))/(k*Temp))))
    return N
x = np.linspace(-10, 10, num = 30000)

```

```python
cd_percents = []
temp_percents = []
percent_ratios = []
percent_ratios_num = []
for i in range(1, 50):
    cd_list = CD_Temp((i*np.sin(x) + 500))
    cd_amp = (max(cd_list) - min(cd_list)) / 2
    cd_avg = max(cd_list) - cd_amp
    cd_wave_percent = (cd_amp/cd_avg) * 100
    cd_percents.append(cd_wave_percent)
    temp_wave_percent = (i/500) * 100
    temp_percents.append(temp_wave_percent)
    percent_ratios.append(str(temp_wave_percent) + ' : ' + str(cd_wave_percent))
    percent_ratios_num.append(temp_wave_percent/cd_wave_percent)
    #print('The ratio of the temperature wave percent to the column density wave percent is ' + str(temp_wave_percent) + ': ' + str(cd_wave_percent))
print(percent_ratios)
print(Average(percent_ratios_num))

```

```python
plt.figure()
plt.plot(cd_percents, temp_percents)
plt.ylabel('Temperature Wave Percent (%)')
plt.xlabel('Column Density Wave Percent (%)')
plt.title('Ratio of Temperature and Column Density Wave Percents')
plt.show()
```
